```python
import pandas as pd
import folium
import math 
```

```python
df_carAccident = pd.read_csv('C:/Users/a/Desktop/lesson/practice01.csv', encoding = 'cp950')
```

```python
df_carAccident.head()
```

```python
m0 = folium.Map([23.5,121],zoom_start = 8.5)
```

```python
for itr in range(len(df_carAccident)):
    latVal = df_carAccident.iloc[itr]['中心點緯度']
    lngVal = df_carAccident.iloc[itr]['中心點經度']
    totalVal = df_carAccident.iloc[itr]['總傷亡人數']
    cRad = math.sqrt(totalVal)*0.5
    information = str(df_carAccident.iloc[itr]['行政區名'])+ '<br>' + 'Injuries:' +str(df_carAccident.iloc[itr]['總受傷人數'])+ '<br>' + 'Deaths:' + str(df_carAccident.iloc[itr]['總死亡人數'])+ '<br>' + 'Total:' + str(df_carAccident.iloc[itr]['總傷亡人數'])
    folium.CircleMarker(

        location = [latVal,lngVal],
        radius = cRad,
        color = '#be395e',
        fill = True,
        fill_opacity = 0.15,
        popup = information,

    ).add_to(m0)
```

```python
for tmpscale in range(2,5):
    
    folium.CircleMarker(

        location = [22.8,122],
        radius = math.sqrt(10**tmpscale)*0.5,
        color = 'red',
        fill = True,
        fill_opacity = 0.15,

    ).add_to(m0)
```
